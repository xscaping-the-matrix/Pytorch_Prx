Numpy to Tensor

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
X_train_tensor.shape, X_test_tensor.shape, y_train_tensor.shape, y_test_tensor.shape
```

Defining the Model

```python
class SimpleNN():

    def __init__(self, X):
        self.weights = torch.randn(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
        self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

    def forward(self, X):
        z = torch.matmul(X, self.weights) + self.bias
        return torch.sigmoid(z)
    def loss_fn(self, y_pred, y):
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
        loss = -(y * torch.log(y_pred) + (1-y) * torch.log(1 - y_pred)).mean()
        return loss
```

```python
lrate = 0.1
epochs = 1000
```

```python
# Create the model
model = SimpleNN(X_train_tensor)
# Train the model
for epoch in range(epochs):
    # Forward pass
    y_pred = model.forward(X_train_tensor)
    # Compute the loss
    loss = model.loss_fn(y_pred, y_train_tensor)
    # Backward Pass
    loss.backward()

    # Update weights and bias
    with torch.no_grad():
        model.weights -= lrate * model.weights.grad
        model.bias -= lrate * model.bias.grad
    # Zero the gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()
    # Print the loss every 10 epochs
    print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}')
```

```python
# model evaluation
with torch.no_grad():
    y_test_pred = model.forward(X_test_tensor)
    #print(y_test_pred.shape)
    y_test_pred = (y_test_pred > 0.5).float()
    #print(y_test_pred.float())
    acc = (y_test_pred == y_test_tensor).float().mean()
    print(acc.item()) 
```
